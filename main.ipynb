```python
import random
import numpy as np
import turtle
import matplotlib.pyplot as plt
import matplotlib.colors as pp
import matplotlib.patches as mpatches
```

# график функции одной преременной

```python
def draw_f(f, segment_start = -20, segment_end = 20, frequency = 700 ):
    plt.subplots()
    x = np.linspace(segment_start, segment_end, frequency)
    plt.plot(x, f(x), color="k")
    plt.show()
```

```python
draw_f(lambda x: np.sin(x/3)*np.cos(3*x)) # кардиограмма
```

# Треугольник серпинского рекурсивно



```python
ax = plt.axes([0,0,1,1])

def draw(p, color_):
    tmp=mpatches.Polygon(np.array(p), color=color_)
    ax.add_artist(tmp)

def middle_point(p1, p2):
    return (p1[0] + p2[0])/2, (p1[1] + p2[1])/2

def sierpinski(p, generation):
    color_ = ['k','g','r', 'c', 'm', 'y', '#006400', '#006400', '#FF69B4', '#0000CD']
    draw(p, color_[generation])
    if generation > 0:
        sierpinski([p[0], middle_point(p[0], p[1]), middle_point(p[0], p[2])], generation-1)
        sierpinski([p[1], middle_point(p[0], p[1]), middle_point(p[1], p[2])], generation-1)
        sierpinski([p[2], middle_point(p[2], p[1]), middle_point(p[0], p[2])], generation-1)

sierpinski([[-20, -10], [0, 20], [20, -10]], 8)
ax.set_xlim(-20, 20)
ax.set_ylim(-20, 20)
plt.show()
```

# Треугольник серпинского хаус


```python
ax = plt.axes([0,0,1,1])
tr=np.array([[-20, -10], [0, 20], [20, -10]])
def draw(p, color_):
    tmp=mpatches.Polygon(np.array(p), color=color_)
    ax.add_artist(tmp)

def middle_point(p1, p2):
    return (p1[0] + p2[0])/2, (p1[1] + p2[1])/2

def draw_sierpinsky(p, generation):
    tmp=plt.scatter(p[0], p[1], color='black')
    ax.add_artist(tmp)
    if generation > 0:
        rand= random.random()
        if rand < 1/3 :
            draw_sierpinsky(middle_point( tr[0], p), generation-1)
        elif rand < 2/3:
            draw_sierpinsky(middle_point( tr[1], p), generation-1)
        else:
            draw_sierpinsky(middle_point( tr[2], p), generation-1)
plt.plot(tr[: , 0], tr[ : , 1], 'k')
draw_sierpinsky([0.4, -3], 600)

plt.show()

```
